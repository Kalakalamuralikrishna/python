```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Global YouTube Statistics.csv", encoding='latin1')

```

```python
print("\nFirst 5 Rows:")
print(df.head())
print(df.columns)

top_subscribers = df[['Youtuber', 'subscribers']].sort_values(
    by='subscribers', ascending=False).head(10)

print("\nTop 10 Channels by Subscribers:")
print(top_subscribers)

top_views = df[['Youtuber', 'video views']].sort_values(
    by='video views', ascending=False).head(10)

print("\nTop 10 Channels by Video Views:")
print(top_views)

country_creators = df['Country'].value_counts().head(10)

print("\nTop 10 Countries by Number of YouTube Creators:")
print(country_creators)

```

```python
plt.figure(figsize=(5, 10))
sns.barplot(x='subscribers', y='Youtuber', data=top_subscribers, palette='dark')
plt.title('Top 10 by Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('Channel')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='video views', y='Youtuber', data=top_views, palette='magma')
plt.title('Top 10 by Video Views')
plt.xlabel('Video Views')
plt.ylabel('Channel')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x=country_creators.values, y=country_creators.index, palette='Blues_d')
plt.title('Top 10 Countries by Number of YouTube Creators')
plt.xlabel('Number of Creators')
plt.ylabel('Country')
plt.show()
```
